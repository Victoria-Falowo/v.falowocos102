```python
import tkinter as tk
from tkinter import messagebox

#Parent Class
class Zenith:
    def mutual_services(self):
        return["Lines of credit", "Investment management and accounts", "Insurance"]
    def unique_services(self):
        return []

#Sub classes
class RetailBanking(Zenith):
    def unique_services(self):
        return ["Retirement and education accounts", "Loans and mortgagges", "Checking and saving"]
    
class GlobalBanking(Zenith):
    def unique_services(self):
        return ["Multi-currency management services and products", "Foreign currency accounts", 
               "Foreign currency credit cards", "Transborder advisory services", "Liquidity management"]
    
class CommercialBanking(Zenith):
    def unique_services(self):
        return ["Advisory services"]
    
#Functions to get services based on division
def get_services(name, division):
    if division == "Retail Banking":
        employee = RetailBanking()
    elif division == "Global Banking":
        employee = GlobalBanking()
    elif division == "Commercial Banking":
        employee = CommercialBanking()
    else:
        messagebox.showerror("Error", "Invalid division selected")
        
    mutual = employee.mutual_services()
    unique = employee.unique_services()
    services = mutual + unique
    services_list = "\n".join(services)
    
    messagebox.showinfo("Services Rendered", f"Employee: {name} \nDivision: {division} \nServices: \n{services_list}")
    
#GUI

def show_services():
    name = name_entry.get()
    division = division_drop.get()
    if not name or not division:
        messagebox.showerror("Error", "Please enter all fields")
        return
    get_services(name, division)


#Main window
root = tk.Tk()
root.title("Zenith Bank Services")
root.geometry("700x700")

#Employee name and label
name_label = tk.Label(root, text = "Employee Name:")
name_label.pack()

name_entry = tk.Entry(root)
name_entry.pack()

#Civision Label
division_label = tk.Label(root, text="Division:")
division_label.pack()

# Dropdown for division
division_drop = tk.StringVar()
division_menu = tk.OptionMenu(root, division_drop, "Retail Banking", "Global Banking", "Commercial Banking")
division_menu.pack()

# Button to get services
button = tk.Button(root, text = "Get Services", command = show_services)
button.pack()


#Run the main event Loop
root.mainloop
```
